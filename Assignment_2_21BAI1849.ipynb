```python
import pandas as pd
import numpy as np
```

#________________ LOADING THE DATASET ________________#

```python
data = pd.read_csv('/content/House Price India.csv')
data
```

# Descriptive Statistics

```python
descriptive_stats = data.describe()
descriptive_stats
```

# Handling missing values

```python
data.fillna(data.mean(), inplace=True)
```

# Performing the Visualizations #

```python
# Univariate Analysis
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for 'number of bedrooms'
plt.figure(figsize=(8, 5))
sns.histplot(data['number of bedrooms'], kde=True)
plt.title('Number of Bedrooms Distribution')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bivariate analysis
# Scatter plot: 'living area' vs. 'Price'
plt.figure(figsize=(8, 5))
sns.scatterplot(x='living area', y='Price', data=data)
plt.title('Scatter Plot: Living Area vs. Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

```python
# Multivariate analysis
# Pair plot for selected columns
columns_to_include = ['living area', 'number of bedrooms',
                      'number of bathrooms','Price']
sns.pairplot(data[columns_to_include])
plt.show()
```
