```python
import numpy as np
import pandas as pd
```

Task 1: Create a pandas dataframe using numpy with 4 features and 4 observations

```python
np.random.seed(44)
```

```python
data = np.random.rand(4, 4)
```

```python
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4'])
df
```

Task 2: Renaming the columns

```python
df.columns = ['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4']
df
```

Task 3: Descriptive stats

```python
descriptive_stats = df.describe()
descriptive_stats
```

Task 4: Checking null values.

```python
# Checking for null values
null_values = df.isnull().sum()

# Finding the data types of the columns
data_types = df.dtypes

# Displaying the null values and data types
print("Null Values:")
print(null_values)
print("\nData Types:")
print(data_types)
```

Task 5: loc and iloc methods

```python
# Display 'Random value 2' and 'Random value 3' columns using .loc
selected_columns_loc = df.loc[:, ['Random value 2', 'Random value 3']]
print(selected_columns_loc)

# Display 'Random value 2' and 'Random value 3' columns using index location
selected_columns_idx = df.iloc[:, [1, 2]]
print(selected_columns_idx)
```
